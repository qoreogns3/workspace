```python
# 주어진 딕셔너리 데이터로 1~2번 문제를 푸세요
student = {
  'name'  :'hong',
  'korScore' : 0,
  'mathScore' : 0
}
```

```python
#1. 위 딕셔너리 데이터에 키보드로 입력한 값으로 국어점수 및 수학점수를 변경
student['korScore'] = int(input('국어점수 : '))
student['mathscore'] = int(input('수학점수 : '))
```

```python
#2. 위 딕셔너리 데이터에 총점(totalScore)를 추가하고, value 에는 국어 및 수학 점수의 합을 저장
student['totalScore'] = student['korScore'] + student['mathscore']
print(student)
```

```python
#3. 아래 딕셔너리 데이터가 매개변수로 주어지면 모든 정수의 합을 리턴하는 함수 선언 맟 호출
scores = {
  'kor' : 70,
  'math' : 80,
  'eng' : 90
}



#함수 선언
def scoresSum(scores) : 
  #매개변수로 전달된 모든 정수의 수만큼 반복
  scorelist = []
  for e in scores : 
    scorelist.append(scores[e])
  return(sum(scorelist))
```

```python
# 3번 문제의 함수를 호출
scoresSum(scores)
```

```python
'''
매개변수로 전달된 딕셔너리에 주어진 key와 value를 추가하는함수를 선언 및 호출
첫번째 매개변수 : 딕셔너리
두번째 매개변수 : 추가할 key
세번째 매개변수 : 추가할 value
'''

# 4번 함수 선언
def addDic(dic = {}, key = '', value = '') : 
  dic[f'{key}'] = value
```

```python
# 4번 함수 호출
addDic(scores, 'eng2', 75)
print(scores)
```

```python
'''
5번
매개변수로 문자열 리스트가 전달되면 리스트의 문자열과 문자열의 길이를 딕셔너리로 리턴하는 함수 선언 및 호출

ex> 매개변수 : ['apple', 'banana', 'orange']
결과 : {
  'apple' : 5
  'banana' : 6
  'orange' : 6
}
'''

#5번 함수 선언
def readList(list = []) :
  dic = {} 
  for e in list :
    dic[f'{e}'] = len(e)
  return dic
```

```python
#5번 함수 호출
readList(['apple', 'banana', 'orange'])
```
