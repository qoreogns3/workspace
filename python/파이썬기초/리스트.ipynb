```python
#리스트(다수의 데이터를 index 번호로 관리하는 자료형)
#변수명에 list는 사용 금지!
list1 = []
list2 = [1, 5.5, 'python']

print(list1)
print(list2)
print(list2[1])
```

```python
#리스트에 데이터 추가하기
list1 = [1,2,3]

#1. append() : 자바스크립트의 push()와 동일한 기능(list의 가장 마지막 요소에 데이터 추가)
list1.append(4) #{1,2,3,4}

#2. insert(위치, 데이터) : 원하는 index에 데이터 추가
list1.insert(2,5) #{1,2,5,3,4}

print(list1)
```

```python
#리스트의 데이터 삭제

#del 명령어 사용
del list1[1] #list1의 1번 요소 데이터를 삭제
print(list1)

#주의사항
#del list1 #list1이라는 변수 자체릉 삭제
#print(list1) 위 코드의 실행으로 변수 자체가 삭제되기 때문에 오류!
```

```python
# 리스트 데이터 수정
list2 = [1, 3.3, 'python']

# 1번 요소를 'hello'로 변경
list2[1] = 'hello'

print(list2)
```

```python
#리스트 합치기
list1 = [1,2,3]
list2 = [4,5,6]

#[1,2,3,4,5,6]
list3 = list1 + list2
print(list3)

list4 = list1 * 3
print(list4)
```

```python
# 리스트의 슬라이싱 연산(리스트의 특정 부분만 추출)
# [시작 index : 마지믹 index]    (마지막 index는 포함되지않음)
list1 = [1,2,3,4,5]
print(list1[1:4]) 

#list1 배열에서 [2,3] 데이터만 같는 새로운 리스트를 생성
list2 = list1[1:3]
print(list2)

list3 = [1,2,3,4,5]
print(list3[2:]) #2번 index부터 끝까지
print(list3[:3]) #처음부터 3번 전까지
print(list3[:])  #전부 출력
print(list3[2:-1])

```

```python
#파이썬에서는 문자열을 리스트로 취급
myName = 'hong'
print(myName[1])

addrList= [
  '스타벅스 삼산지점 052-1111-2222',
  '스타벅스 신정동 2호점 052-2222-3333',
  '스타벅스 울산 남구점 052-3333-4444'
]

#스타벅스 매정 정보에서 연락처 정보만 추출
print(addrList[0][10:])
print(addrList[1][13:])
print(addrList[2][-13:])
```
