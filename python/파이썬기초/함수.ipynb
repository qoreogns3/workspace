```python
# 파이썬은 매개변수 사용 시 매개변수 이름을 자주 사용한다.
def tellMe(name, age, addr):
  print(f'이름은 {name}, 나이는 {age}, 주소는 {addr}')
```

```python
tellMe('kim', 20, 'ulsan')

tellMe(name = 'kim', age = 20, addr = 'ulsan')
tellMe(age = 20, name = 'kim',  addr = 'ulsan')
```
