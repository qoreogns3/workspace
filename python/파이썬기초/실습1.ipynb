```python
#1. 1부터 키보드로 입력받은 수까지의 합을 출력
total = 0
num = int(input('정수입력 : '))
for e in range(1, num+1):
  total = total + e
print(total)
```

```python
#2. 키보드로 입력받은 단수의 구구단을 출력
num = int(input('정수입력 : '))
for e in range(1, 10) : 
  print(f'{num} * {e} = {num*e}')
```

```python
#3. 아래 제시된 리스트의 요소 중 가장 큰 수를 출력
list1 = [3,1,7,10,5]

print(max(list1))
```

```python
#4. 아래 제시된 리스트의 모든 요소의 합을 출력
list2 = [1,3,5]
list3 = [10,15,20]

print(sum(list2+list3))
```

```python
#5. 아래의 데이터에서 't'의 갯수를 구하시오
data= ['mother', 'father', 'brother']
cnt = 0
for e in data : 
  for i in e : 
    if i == 't' : 
      cnt = cnt + 1 
print(cnt)
```
