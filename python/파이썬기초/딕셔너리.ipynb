```python
#Dictionary(딕셔너리) : 자바스크립트의 객체

#딕셔너리 데이터 생성
dic1 = {} #빈 딕셔너리 생성
dic2 = {
  '학번' : 3,
  '이름' : 'kim',
  '점수' : 80
}

print(dic2)
print(dic2['학번']) #특정 키값에 대한 value만 출력
```

```python
#데이터 추가 및 변경
phone = {
  'modelName' : 's20',
  'price' : 1000
}

#데이터 추가
#없는 key에 값을 저장하면 새로운 key가 추가됨
phone['color'] = 'red' #color라는 key를 생성하고 value에 'red' 문자열을 저장

#데이터 수정
phone['price'] = 2000

print(phone)
```

```python
#딕셔너리 데이터 삭제 : del 명령어 사용
del phone['color']
print(phone)
```

```python
# 연습
family = {
  '이름' : '홍길동',
  '나이' : 20,
  '취미' : ['운동', '독서', '등산'],
  '가족관계' : [
    {
      '관계' : '부',
      '이름' : '홍아빠'
    }, 
    {
      '관계' : '모',
      '이름' : '홍엄마'
    }
  ]
}
```

```python
#모든 취미 출력
print(family['취미'])
#취미의 갯수 출력
print(len(family['취미'])) #len() 매개변수로 리스트가 전달되면 리스트에 저장된 데이터 수를 변환
#마지막 취미를 출력
print(family['취미'][-1])
#아버지의 이름 출력
print(family['가족관계'][0]['이름'])
#어머니 정보에 나이 정보를 추가(나이는 60으로 설정)
family['가족관계'][1]['나이'] = 60
print(family)
```
