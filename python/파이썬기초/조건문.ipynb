```python
num1 = 5 

# 파이썬 조건문의 실행 영역은 들여쓰기로 범위를 구분한다.
if num1 > 3 : 
  print('num1은 3보다 큽니다')
else : 
  print('num1은 3보다 작습니다.')  
```

```python
num2 = 3

#num2 변수의 값이 3보다 크면 1, 3보다 작으면 2, 3이면 0을 출력
if num2 > 3 :
  print(1)
elif num1 < 3 :
  print(2)
else :
  print(3)
```

```python
# input() : 키보드로 데이터 입력받기
# input으로 입력받은 데이터는 모두 문자열로 인식
inputedData = input('정수 입력 : ')
print(inputedData) 
```

```python
#키보드로 입력받은 정수가 짝수인지 홀수인지 판별 및 출력
num = int(input())
if num % 2 == 0 :
  print('짝수')
else :
  print('홀수')
```

```python
#조건이 두개인 경우 and, or 를 사용
if 10 > 3 and 5 > 2 :
  print(1)
```
