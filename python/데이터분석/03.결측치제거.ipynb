```python
import pandas as pd
```

```python
# 참 : True, 거짓 : False, 널데이터 : None

data = {
    '학번': ['2024001', '2024002', '2024003', '2024004', '2024005',
             '2024006', '2024007', '2024008', '2024009', '2024010'],
    '이름': ['김민준', '이서연', '박지호', '최수아', '정우진',
             '강하은', '윤도현', '임지유', '한준서', '오서현'],
    '학과': ['컴퓨터공학', '정보통신', '컴퓨터공학', None, '정보통신',
             '컴퓨터공학', None, '정보통신', '컴퓨터공학', '소프트웨어'],
    '학년': [1, 2, 1, 3, 2, 1, 3, 2, 1, 3],
    '파이썬점수': [85, 92, None, 95, 88, 73, 90, None, 76, 94],
    '자바점수': [90, None, 82, 88, 92, None, 95, 78, 80, 87],
    '출석률': [95.0, 98.0, 85.0, None, 90.0, 78.0, None, 88.0, 82.0, 96.0]
}

df = pd.DataFrame(data)
df
```

```python
#결측치 확인하기
df.isnull() #각 데이터를 True, False로 결측치 표현

#결측치가 있는 행만 보기
#axis=0 -> 행, axis=1 -> 열
df[ df.isnull().any(axis = 1) ]

#각 컬럼별 결측치 갯수 확인
df.isnull().sum()
```

```python
#df의 내용을 test_df로 복사(df의 내용은 기존 내용 보존을 위해)
test_df = df.copy()
```

```python
#결측치 제거
#dropna(axis='index', how='any')
#axis='index' 결측치가 있는 행을 삭제(default)
#axis='columns' 결측치가 있는 열을 삭제
#how='any' 결측치가 하나라도 존재하면 삭제(default)
#how='all' 모든 데이터가 결측치일때 삭제

#dropna() 함수를 사용하면 결측치가 제거된 결과 데이터를 리턴, 원본 데이터는 유지
result1_df = test_df.dropna()
result1_df

#test_df를 출력하면 원본 데이터 유지
test_df

#결측치가 하나라도 존재하는 열을 삭제
result2_df = test_df.dropna(axis='columns', how='any')
result2_df

#--특정 컬럼에 결측치가 존재하는 행만 제거--#
#ex> 파이썬 컬럼에 결측치가 존재하는 행 제거
result3_df = test_df.dropna(subset=['파이썬점수']) 
result3_df

#ex> 리스트에 나열한 컬럼 중 하나라도 결측치가 포함되어있으면 삭제
result4_df = test_df.dropna(subset=['파이썬점수', '자바점수'])
result4_df
```

```python
#결측치를 특저 데이터로 대체

#결측치를 고정값으로 대체
#ex> 학과의 결측치를 '미정'으로 채우기
test_df['학과'] = test_df['학과'].fillna('미정')
test_df

#데이터에서 숫자 데이터의 통계 데이터 확인
test_df.describe()

#특정 컬럼의 통계 데이터확인
#파이썬점수의 통계
test_df['파이썬점수'].mean() #평균
test_df['파이썬점수'].max() #최댓값
test_df['파이썬점수'].min() #최솟값
test_df['파이썬점수'].median() #중앙값

#ex> 파이썬점수의 결측치를 파이썬 점수의 평균으로 채우기
test_df['파이썬점수'] = test_df['파이썬점수'].fillna(test_df['파이썬점수'].mean())
test_df
```

```python
#1. 위 test_data를 pandas의 DataFrame으로 만드세요
df = pd.DataFrame(test_data)
```

```python
#4. 결측치가 하나라도 있는 행을 모두 제거하고, 몇개의 행이 제거됐는지 출력
#참고 > len(DataFrame) : 데이터의 갯수
len(df) - len(df.dropna())
```

```python
#5. 원본데이터에서 연봉과 성과점수 중 결측치가 포함된 행을 제거
result1_df = df.dropna(subset=['연봉', '성과점수'])
result1_df
```

```python
#6. 아래 요구사항에 맞게 결측치를 채우세요
# 1) 부서의 결측치는 '미배정'으로 채움
df['부서'] = df['부서'].fillna('미배정')
# 2) 연봉의 결측치는 연봉 평균으로 채움
df['연봉'] = df['연봉'].fillna(df['연봉'].mean())
# 3) 성과점수 결측치는 중앙값으로 채움
df['성과점수'] = df['성과점수'].fillna(df['성과점수'].median())

df
```
