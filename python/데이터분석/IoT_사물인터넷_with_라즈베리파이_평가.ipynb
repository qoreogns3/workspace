### 문제 3.
df 데이터를 아래 기준으로 결측치를 처리하세요. 처리된 결과는 **df에 저장**되어야 합니다.

1. `도서분류` 컬럼 → **'미분류'** 로 채우기
2. `할인율` 컬럼 → **0** 으로 채우기
3. `판매가격` 컬럼 → **평균값** 으로 채우기
4. `수량` 컬럼 → **중앙값** 으로 채우기
5. `평점` 컬럼 → **평균값** 으로 채우기 (소수점 둘째 자리)

```python
# 문제 3번 풀이 작성
df['도서분류'] = df['도서분류'].fillna('미분류')
df['할인율'] = df['할인율'].fillna(0)
df['판매가격'] = df['판매가격'].fillna(df['판매가격'].mean())
df['수량'] = df['수량'].fillna(df['수량'].median())
df['평점'] = df['평점'].fillna(round(df['평점'].mean(), 2))

df
```

### 문제 9. ⭐ (심화)

### 문제 10. ⭐ (심화)
df 데이터를 활용하여 아래 조건을 모두 만족하는 결과를 만들고 **엑셀 파일로 저장** 하세요.

1. **'총판매액'** 컬럼을 추가하세요. (`판매가격` × `수량`)
2. **출판사별** 로 총판매액의 합계를 구하고 컬럼명을 **'출판사별_합계'** 로 변경하세요.
3. **출판사별_합계 기준 내림차순** 으로 정렬하고 인덱스를 초기화하세요.
4. 합계가 가장 높은 출판사에는 **'★ 1위'**, 나머지는 **'　'** 를 값으로 갖는 **'비고'** 컬럼을 추가하세요.
5. 완성된 데이터를 **'출판사별_판매분석.xlsx'** 파일로 저장하세요. (인덱스 저장 안 함)

```python
# 문제 10번 풀이 작성
df1 = df.groupby('출판사')['총판매액'].agg(['sum']).rename(columns={'sum' : '출판사별_합계'}).sort_values('출판사별_합계', ascending=False).reset_index()

df1['비고'] = ''
df1['비고'] = df1['비고'].mask(df1['출판사별_합계'] == max(df1['출판사별_합계']), '★ 1위')

df1.to_excel('출판사별_판매분석.xlsx', index=False)
                  
```
