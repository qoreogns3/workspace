```python
import pandas as pd
```

```python
data = {
    '학번': ['2024001', '2024002', '2024003', '2024004', '2024005',
             '2024006', '2024007', '2024008', '2024009', '2024010'],
    '이름': ['김민준', '이서연', '박지호', '최수아', '정우진',
             '강하은', '윤도현', '임지유', '한준서', '오서현'],
    '학과': ['컴퓨터공학', '정보통신', '컴퓨터공학', '소프트웨어', '정보통신',
             '컴퓨터공학', '소프트웨어', '정보통신', '컴퓨터공학', '소프트웨어'],
    '학년': [1, 2, 1, 3, 2, 1, 3, 2, 1, 3],
    '파이썬점수': [85, 92, 78, 95, 88, 73, 90, 82, 76, 94],
    '자바점수': [90, 85, 82, 88, 92, 70, 95, 78, 80, 87],
    '출석률': [95.0, 98.0, 85.0, 100.0, 90.0, 78.0, 92.0, 88.0, 82.0, 96.0]
}

df = pd.DataFrame(data)
df
```

```python
# 새 컬럼 및 데이터 추가
#ex> '총점수' 컬럼 생성 및 총점 데이터 추가
df['총점수'] = df['파이썬점수'] + df['자바점수']
df
```

```python
#출석률을 소수점 비율로 변경 (95.0 -> 0.95)
df['출석률'] = df['출석률'] / 100
df

```

```python
#조건에 맞는 데이터 변경(파이썬 점수에 따른 등급 정보)
#1. 등급 컬럼 생성
df['등급']='C'
df

#2. '파이썬점수'가 90점 이상이면 등급을 'A'로 변경
#mask(조건, 변경할 값)
df['등급'] = df['등급'].mask(df['파이썬점수']>=90,'A')
df

#3. '파이썬점수가' 가 80~89점이면 등급을 'B'로 변경
df['등급'] = df['등급'].mask((df['파이썬점수']<=89) & (df['파이썬점수']>=80),'B')
df
```

```python
#값 치환
#ex> 정보통신 -> IT, 소프트웨어 -> SW
#replace('A', 'B') -> 'A'값을 'B'로 변경
#replace({'A' : 'B', 'C' : 'D'}) -> 'A'를 'B'로, 'C'를 'D'로
df['학과'] = df['학과'].replace({'정보통신' : 'IT', '소프트웨어' : 'SW'})
df
```

```python
#컬럼명 변경
#파이썬점수 -> python, 자바점수 -> java
df = df.rename(columns={'파이썬점수' : 'python', '자바점수' : 'java'})
df
```

```python
# ------------------------- 실습 ----------------------------#
```

```python
#1. 평균점수 컬럼 생성, 해당 컬럼에는 파이썬과 자바점수의 평균값을 저장(소수점 둘째자리)
#소수점 : round()
df['평균'] = round((df['python']+df['java'])/2, 2)
df
```

```python
#2. '합격여부' 컬럼 생성, 평균 점수가 80점 이상이면 '합격', 미만이면 '불합격'
df['합격여부'] = '합격'
df

df['합격여부'] = df['합격여부'].mask(df['평균'] >= 80, '합격')
df
df['합격여부'] = df['합격여부'].mask(df['평균'] < 80, '불합격')
df
```

```python
#3. '합격여부' 컬럼명을 'pass-yn'으로 변경
df = df.rename(columns={'합격여부':'pass-yn'})
df
```
