```python
#단일 컬럼의 정보만 출력
df['학과']

#여러 컬럼 선택
#ex> 이름, 학과 정보만 선택
df[['이름', '학과']]
```

```python
#조건 필터링

#단일 조건 필터링
#ex> 파이썬 점수가 90점 이상인 학생 정보
df['파이썬점수'] >= 90 #각 행 데이터가 조건에 부합하는지 boolean타입 결과를 반환

#정답 코드
df[df['파이썬점수'] >= 90]

#ex> 학과가 '컴퓨터공학'인 학생 정보 조회
df[df['학과'] == '컴퓨터공학']
```

```python
#다중 조건 필터링(and : &, or : |)
#ex> 파이썬 점수가 80점 이상이면서 학과가 '컴퓨터공학'인 학생 정보 조회

#조건1
df['파이썬점수']>=80
df['학과']=='컴퓨터공학'

#정답코드
df[(df['파이썬점수']>=80) & (df['학과']=='컴퓨터공학')]
```

```python
#여러 값 중 하나의 값에 해당하는 데이터 조회
#ex>학과가 '컴퓨터공학'이거나 '소프트웨어'인 학생 정보 조회

#다중 조건을 이용한 풀이
df[(df['학과'] == '컴퓨터공학') | (df['학과'] == '소프트웨어')]

#isin() 함수를 이용한 출력
df[df['학과'].isin(['컴퓨터공학', '소프트웨어'])]
```

```python
#조건에 맞는 특정 컬럼 데이터만 조회
#파이썬 점수가 90점 이상인 학생들의 이름, 학과, 파이썬 점수를 조회
# df[조건][컬럼명]
df[(df['파이썬점수']>=90)][['이름','학과','파이썬점수']]
```

```python
#1. 이름, 출석률 커럼 데이터 조회
df[['이름', '출석률']]
#2. 출석률이 90이상인 학생 정보를 조회
df[df['출석률'] >= 90]
#3. 학년이 2학년이면서 자바 점수가 80점 이상인 학생의 이름, 학년, 자바점수를 조회
df[(df['학년']==2) & (df['자바점수']>=80)][['이름','학년','자바점수']]
#4. 학과가 '정보통신' 혹은 '소프트웨어'인 학생 정보 출력
df[df['학과'].isin(['정보통신', '소프트웨어'])]
```
