```python
import pandas as pd
```

```python
data = {
    '학번': ['2024001', '2024002', '2024003', '2024004', '2024005',
             '2024006', '2024007', '2024008', '2024009', '2024010'],
    '이름': ['김민준', '이서연', '박지호', '최수아', '정우진',
             '강하은', '윤도현', '임지유', '한준서', '오서현'],
    '학과': ['컴퓨터공학', '정보통신', '컴퓨터공학', '소프트웨어', '정보통신',
             '컴퓨터공학', '소프트웨어', '정보통신', '컴퓨터공학', '소프트웨어'],
    '학년': [1, 2, 1, 3, 2, 1, 3, 2, 1, 3],
    '파이썬점수': [85, 92, 78, 95, 88, 73, 90, 82, 76, 94],
    '자바점수': [90, 85, 82, 88, 92, 70, 95, 78, 80, 87],
    '출석률': [95.0, 98.0, 85.0, 100.0, 90.0, 78.0, 92.0, 88.0, 82.0, 96.0]
}

df = pd.DataFrame(data)
df
```

```python
#집계함수
#df.groupby(그룹화할 기준 컬럼)[결과컬럼].집계데이터
#집계데이터 : mean(), max(), min(), count(), median()

#학과별 파이썬점수의 평균
df.groupby('학과')['파이썬점수'].mean()
```

```python
#학과별 학생수
df.groupby('학과')['학번'].count()
```

```python
#여러 기준으로 그룹핑
#학과 및 학년으로 그룹핑후 파이썬점수의 평균을 조회
df.groupby(['학과', '학년'])['파이썬점수'].mean()
```

```python
#여러 집계 데이터 함께 조회
#학년별 파이썬 점수의 평균, 최대, 최소 점수를 조회
df.groupby('학년')['파이썬점수'].agg(['mean','max','min'])
```

```python
#학과별로 파이썬점수와 자바점수의 평균
df.groupby('학년')[['파이썬점수', '자바점수']].mean()
```

```python
#1. 학과별 평균 출석률 조회
df.groupby('학과')['출석률'].mean()
```

```python
#2. 학년별 파이썬과 자바 점수의 최고점, 최저점을 조회
df.groupby('학년')[['파이썬점수', '자바점수']].agg(['max','min'])
```

```python
#3. 학과별 학생수를 조회, 조회된 데이터를 'group_result.xlsx' 파일로 저장
# 원래 그룹화한 기준컬럼이 index가 됨
df.groupby('학과')['학번'].count().reset_index().to_excel('group_result.xlsx', index=False)

#reset_index() = index를 컬럼으로 변경

```

```python
new_df = pd.read_excel('group_result.xlsx')
new_df
```

```python
#정렬
#df.sort_values(정렬기준컬럼, ascending = True)

#파이썬 점수 기준 내림차순 정렬
df.sort_values('파이썬점수', ascending=False)

#먼저 학과 기준 오름차순 정렬 후 자바 점수 기준 내림차순 정렬
df.sort_values(['학과', '자바점수'], ascending=[True, False])

#정렬 후 index를 재 정렬
#reset_index() : 인덱스를 컬럼으로 변경 + 새로운 인덱스 부여
#reset_index(drop = True) : 기존의 인덱스는 버리고 새로운 인덱스를 부여
df.sort_values(['학과', '자바점수'], ascending=[True, False]).reset_index(drop=True)
```

```python
#1. 출석률이 높은 순으로 데이터를 정렬 후, 상위 3명의 이름과 출석률 조회
df.sort_values('출석률', ascending=False)[['이름', '출석률']].reset_index(drop=True).head(3)
```

```python
#2. df에서 학과기준 오름차순 정렬 후 자바 점수기준 내림차순 정렬
df.sort_values(['학과', '자바점수'],ascending=[True, False])
```

```python
#3. 파이썬과 자바점수의 평균점수를 기준으로, 평점 점수 상위 5명의 데이터를 'top5.xlsx'파일로 저장

# 1) '평균'컬럼 생성 후 해당 컬럼에 자바와 파이썬 점수의 평균 데이터 저장(소숫점 첫째 자리)
df['평균'] = round((df['파이썬점수'] + df['자바점수'])/2, 1)
df
# 2) 평균점수 기준 내림차순 정렬 후 상위 5개의 데이터를 액셀로 저장
df.sort_values('평균', ascending=False).head(5).to_excel('top5.xlsx', index=False)

pd.read_excel('top5.xlsx')

```
