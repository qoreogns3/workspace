문제 1. 공유폴더에 제공한 'order.xlsx' 파일을 불러와 DataFrame 데이터를 'df'로 생성

```python
#문제 1번 풀이 작성
df = pd.read_excel('order.xlsx')
df
```

문제 2. 컬럼별 결측치의 갯수를 파악하여 결측치의 갯수가 가장 많은 컬럼명을 변수 '답_02'에 저장하세요.  
ex> 답_02 = '컬럼1'  
ex> 답_02 = '컬럼1, 컬럼2'

```python
#문제 2번 풀이 작성
df.isnull().sum()
답_02 = '평점'
```

문제 3. df 데이터를 아래 기준으로 결측치를 처리하세요. 처리된 결과는 df에 저장되어야 합니다.  
1) 카테고리 컬럼 -> '미분류'로 채우기  
2) 한인율 -> 0으로 채우기  
3) 판매금액 -> 평균값으로 채우기  
4) 수량 -> 중앙값으로 채우기
5) 평점 -> 평균값으로 채우기(소수점 둘째 자리)

```python
#문제 3번 풀이 작성
df['카테고리'] = df['카테고리'].fillna('미분류')
df['할인율'] = df['할인율'].fillna(0)
df['판매금액'] = df['판매금액'].fillna(df['판매금액'].mean())
df['수량'] = df['수량'].fillna(df['수량'].median())
df['평점'] = df['평점'].fillna(round(df['평점'].mean(), 2))

df
```

문제 4. df에서 '카테고리'와 '지역' 컬럼만 선택하여 출력하세요.

```python
#문제 4번 풀이 작성
df[['카테고리', '지역']]
```

문제 5. df에서 지역이 '서울'이거나 '부산'인 주문 중 평점이 4.0 이상이 주문 정보를 출력하세요.

```python
#문제 5번 풀이 작성
df[(df['지역'].isin(['서울', '부산'])) & (df['평점'] >= 4.0)]
```

문제 6. df 데이터를 아래 요구사항에 맞게 데이터를 변경하세요.  
1) '판매금액'과 '수량'을 곱한 '총판매액' 컬럼을 추가하세요
2) '지역'컬럼에서 '대구'를 '대구광역시'로 변경하세요.

```python
#문제 6번 풀이 작성
df['총판매액'] = df['판매금액']*df['수량']
df
```

문제 7. df 데이터에서 카테고리별 판매금액의 합계를 출력하시오.

```python
#문제 7번 풀이 작성
df.groupby('카테고리')['판매금액'].sum()
```

```python
#문제 8번 풀이 작성
df.sort_values('판매금액', ascending=False).head(5)[['고객명', '판매금액']].reset_index(drop = True)
```
